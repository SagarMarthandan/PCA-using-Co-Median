```python
yeast = pd.read_csv('seeds_dataset.txt', sep="\s+", index_col=False)
yeast.reset_index(drop=True, inplace=True)
yeast_columns = yeast.columns
print(yeast)
categorical_features = ['wheat_class']
numerical_features = ['area_A', 'perimeter_P', 'compactness_C', 'length_of_kernel',
       'width_of_kernel', 'asymmetry_coefficient', 'length_of_kernel_groove']
cat_data = yeast[categorical_features]
num_data = yeast[numerical_features]

scaler = MinMaxScaler(feature_range=(-1, 1))
x = pd.DataFrame(scaler.fit_transform(yeast[numerical_features]))
x.columns = numerical_features
x_comb = pd.concat([cat_data, x], axis=1)
print(f'\n----> SCALED DATASET : \n\n{x_comb}')
s_row, s_col = x.shape

print(f'\n*\tSamples : {s_row}, Dimensions : {len(yeast.columns)}')

#sns.pairplot(data=yeast, hue='wheat_class', height=2, palette='Set1')
#plt.plot()
```
